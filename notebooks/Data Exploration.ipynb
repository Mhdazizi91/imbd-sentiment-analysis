## Prepare Data

```python
train_data = pd.read_csv('train.csv', header=None)
test_data = pd.read_csv('test.csv', header=None)
```

```python
# get text
text = ' '.join(train_data[0]) # join all reviews in training set
# text = train_data[0][0] # first review sample

# limit word count
wordcount = 20000

# stop words
stopwords = set(STOPWORDS)
stopwords.add("br")

# setup word cloud
wc = WordCloud(scale=3, background_color="white", max_words=wordcount, stopwords=stopwords)

# generate word cloud
wc.generate(text)

# store to file
wc.to_file("wordcloud.png")

# show
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.figure()
plt.axis("off")
plt.show()
```
